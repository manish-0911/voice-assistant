```python
from tkinter import *
from PIL import ImageTk,Image
import pyttsx3  
import speech_recognition as sr
import os
import pywhatkit
import pyjokes
import datetime
import wikipedia
import webbrowser

#win = Tk()
#win.title('wanda')
#win.geometry('700x650')

def fxn():


    engine = pyttsx3.init("sapi5")
    voices = engine.getProperty("voices")

    engine.setProperty("voice", voices[1].id)


    def speak(audio):
        engine.say(audio)
        engine.runAndWait()

    def wishMe():
        hour = int(datetime.datetime.now().hour)
        if hour>=0 and hour<12:
            speak("Good Morning!")

        elif hour>=12 and hour<18:
            speak("Good Afternoon")

        else:
            speak("Good evening")

        speak("I am Eva How may I assist you")

    def takeCommand():

        r = sr.Recognizer()
        with sr.Microphone() as source:
            print('Listening...')
            r.pause_threshold = 1
            audio = r.listen(source)

        try:
            print('Recognizing')
            query = r.recognize_google(audio, language='en-in')
            print(f"User said: {query}\n")

        except Exception as e:

            print("Say that again please...")
            speak("could you say that again please")
            return "None"
        return query
    
    if __name__ == "__main__":
        wishMe()
        while True:
            query = takeCommand().lower()


        #logic for exicuting tasks based on query
            if 'wikipedia' in query:
                speak('Searching Wikipedia...')
                query = query.replace("wikipedia","")
                results = wikipedia.summary(query, sentences=2)
                speak("According to wikipedia")
                print(results)
                speak(results)
                #['text']=str(query)

            elif 'open notepad' in query:
                npath = "C:\\Windows\\System32\\notepad.exe"
                os.startfile(npath)
                speak('Opening notepad')

            elif 'open code' in query:
                cpath = "C:\\Users\\Manish\\AppData\\Local\\Programs\\Microsoft VS Code\\Code.exe"
                os.startfile(cpath)
                speak('opening vs code')

            elif 'open command prompt' in query or 'open cmd' in query:
                os.system("start cmd")
                speak("Opening command prompt")

            elif 'open youtube' in query:
                webbrowser.open("youtube.com")
                speak("opening youtube")

            elif 'open google' in query:
                webbrowser.open("google.com")
                speak("opening google")

            elif 'search' in query:
                speak=("what do you want to search")
                url = 'https://google.com/search?q='
                webbrowser.get().open(url)
                speak('Here is what i found' + url)

            elif 'open hotstar' in query:
                webbrowser.open("hotstar.com")
                speak("opening hotstar")

            elif 'play' in query:
                song = query.replace("play","")  
                speak("playing" + song + "on youtube")
                pywhatkit.playonyt(song)

            elif 'the time' in query:
                strTime = datetime.datetime.now().strftime("%H:%M:%S")
                speak(f"The time is {strTime}")

            elif 'single' in query:
                speak('I am in relationship with wifi')

            elif ' joke' in query:
                speak(pyjokes.get_joke(language="en",category="all"))

            elif 'how can you help me' in query:
                speak("I am a virtual assistant of your device. I can open few of the popular sites for you, Play some music to relax you, and can surf though wikipedia")

            elif 'what else can you do for me' in query:
                speak("what do you want me to do")


win = Tk()
win.title('Virtual Assistant')
win.geometry('730x512')
frame=Frame(win,bg='red')

img=ImageTk.PhotoImage(Image.open('eva.jpg'))
panel=Label(win, image=img)
panel.pack(side='right',fill='both',expand='no')
panel.pack(fill='both',expand='yes')


userText=StringVar()
b1=Button(win, text='Run',font=('railways',50,'bold'),bg='#5DA7DB',fg='white',command=fxn).pack(fill='x', expand='no')

userText.set('Your Virtual Assistant')

top=Message( textvariable=userText,bg='#2B3A55',fg='#CE7777')
top.config(font=("Century Gothic",20,'bold'))
top.pack(side='top',fill='both',expand='yes')

#b1=Button(win, text='Run',font=('railways',10,'bold'),bg='red',fg='white',command=fxn).pack(fill='x', expand='no')
b2=Button(win, text='Stop',font=('railways',50,'bold'),bg='#8EC3B0',fg='#153462',command=win.destroy).pack(fill='x', expand='no')


win.mainloop()
```
